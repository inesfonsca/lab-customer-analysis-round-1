```python
# Standardize header names.
cols = [col.lower() for col in file1.columns]
ls = [col.lower() for col in file2.columns]
cols3 = [col.lower() for col in file3.columns]
```

```python
file1.columns = cols
file2.columns = ls
file3.columns = cols3
```

```python
file2 = file2[['customer', 'st', 'gender', 'education', 'customer lifetime value',
               'income', 'monthly premium auto', 'number of open complaints', 'policy type',
               'vehicle class', 'total claim amount']]

file3 = file3[['customer', 'st', 'gender', 'education', 'customer lifetime value',
               'income', 'monthly premium auto', 'number of open complaints', 'policy type',
               'vehicle class', 'total claim amount']]
```

```python
# Concatenate the three dataframes

data = pd.concat ([file1, file2, file3])
```

```python
# Which columns are numerical?
data._get_numeric_data()
```

```python
# Which columns are categorical?
data.select_dtypes('object')
```

```python
# Data cleaning

data.drop_duplicates()
data.drop(['education', 'number of open complaints'], axis=1)
```

```python
# Replace all the '%'' for ''
data['customer lifetime value'] = data['customer lifetime value'].str.replace(r'%', '')
```

```python
data
```

```python
# Converts the column to numerical float and * all values by 100
data['customer lifetime value'] = data['customer lifetime value'].astype('float')*100
```
